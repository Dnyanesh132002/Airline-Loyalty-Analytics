```python
# Loyalty card distribution
sns.countplot(x='Loyalty Card', data=loyalty)
plt.title("Loyalty Card Distribution")
plt.show()
```

```python
# Gender distribution
sns.countplot(x='Gender', data=loyalty)
plt.title("Gender Distribution")
plt.show()
```

```python
# Monthly flight trend
monthly_flights = flight.groupby('Month')['Total Flights'].sum()

monthly_flights.plot()
plt.title("Monthly Flight Trend")
plt.xlabel("Month")
plt.ylabel("Flights")
plt.show()
```

```python
# CLV distribution
sns.histplot(loyalty['CLV'], bins=30)
plt.title("CLV Distribution")
plt.show()
```

```python
# Salary distribution
sns.histplot(loyalty['Salary'], bins=30)
plt.title("Salary Distribution")
plt.show()
```

```python
# Merge datasets
merged_df = pd.merge(
    loyalty,
    flight,
    on='Loyalty Number',
    how='inner'
)
```

```python
# Flights by loyalty card
merged_df.groupby('Loyalty Card')['Total Flights'].sum().plot(kind='bar')

plt.title("Flights by Loyalty Card")
plt.show()
```

```python
# Points earned by customers
merged_df.groupby('Gender')['Points Accumulated'].sum().plot(kind='bar')

plt.title("Points by Gender")
plt.show()
```

```python
# To check avg CLV byb education
merged_df.groupby('Education')['CLV'].mean().plot(kind='bar')

plt.title("Average CLV by Education")
plt.show()
```

```python
# Check avg salary by loyalty card
merged_df.groupby('Loyalty Card')['Salary'].mean().plot(kind='bar')

plt.title("Average Salary by Loyalty Card")
plt.show()
```
